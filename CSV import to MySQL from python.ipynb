```python
try:
    conn = msql.connect(host='localhost',
                           database='Phonepe_Pulse', user='root',
                           password='Kowshika@153')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        
        cursor.execute('DROP TABLE IF EXISTS AggTransByStates;')
       
        cursor.execute("CREATE TABLE AggTransByStates\
                       (State varchar(100),\
                        Year int,\
                        Quater varchar(5),\
                        Transacion_type varchar(50),\
                        Transacion_count int,\
                        Transacion_amount float(50,3))")
                       
        print("AggTransByStates table is created....")
        for i,row in AggTransByStates.iterrows():
            sql = "INSERT INTO Phonepe_Pulse.AggTransByStates VALUES (%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))                        
            conn.commit()
        print("AggTransByStates values are inserted to MySQL....")
except Error as e:
    print("Error while connecting to MySQL", e)
```
